# Data cleaning and preparing

```python
movies_df["date_added"]=movies_df["date_added"].str.replace(",", "").str.strip() ## Strip removes spaces front and back
movies_df["date_added"]=pd.to_datetime(movies_df["date_added"], format="%B %d %Y")

movies_df["director"]=movies_df["director"].fillna("Unknown")
movies_df["cast"]=movies_df["cast"].fillna("Unknown")
movies_df["country"]=movies_df["country"].fillna("Unknown")


movies_df["Season"]=""

columns_name=["show_id", "type", "title", 
              "director", "cast", "country",
              "date_added", "release_year", "rating",
              "duration", "Season", "listed_in", "description"]
movies_df=movies_df.reindex(columns=columns_name)

movies_df["Season"] = movies_df[movies_df["duration"].str.contains("Season")]["duration"] ## Separating Season from duration

## Make a Season Column an Integer

movies_df["Season"]=movies_df["Season"].fillna("0")
movies_df["Season"]=movies_df["Season"].str.replace("Season", "").str.replace("s", "")
movies_df["Season"]=movies_df["Season"].astype(str).astype(int)

## Make a duration column an Integer
movies_df["duration"]=movies_df.duration.str.replace('^(\d+)(.Seasons*)$', "0") ## Remove The Season from the column
movies_df["duration"]=movies_df["duration"].str.replace(" min", "") ## Remove min
movies_df["duration"]=movies_df["duration"].astype(int) ## Convert to Integer



movies_df
```

# Q.3 What are the different types of shows Netflix have?

# Q.4 Visualize different rating of TV show and Movies  

```python
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(x="rating", data=movies_df, palette="Set2", order=movies_df['rating'].value_counts().index[0:15])
```

# Q.5 Visualize the year wise analysis of the Netflix Shows? 

```python
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(y="release_year", data=movies_df, palette="Set2", order=movies_df['release_year'].value_counts().index[0:15])
```

# Q.6 Frequency of movies released in different year 

```python
plt.figure(figsize=(16,8))
movies_df[movies_df["type"]=="Movie"]["release_year"].value_counts()[:20].plot(kind="bar",color="green")
plt.title("Frequency of Movies which were released in different years and are available on Netflix");
```

```python
plt.figure(figsize=(12,6))
movies_df[movies_df["type"]=="TV Show"]["release_year"].value_counts()[:20].plot(kind="bar",color="yellow")
plt.title("Frequency of TV shows which were released in different years and are available on Netflix")
```
